```python
import numpy as np
import pandas as pd

def calculate_phi(Y):
     return Y.mean()

def calculate_sigma(X, Y, mu0, mu1):
    m = len(Y)
    n= len(X[0])
    sqr_sum = np.zeros((n,n))
    for i in range(0,m):
        xi = X[i]
        yi = Y[i]
        if yi==1:
            sqr= (xi-mu1)*np.transpose((xi-mu1))
        else :
            sqr= (xi-mu0)*np.transpose((xi-mu0))
        sqr_sum = sqr_sum + sqr
 
    return sqr_sum/m



def calculate_mu1(X, Y):
    m = len(Y)
    n=len(X[0])
    y_pos=0
    conditional_sum_x = np.zeros((1,n))
    
    for i in range(0,m) :
        xi = X[i]
        yi = Y[i]
        if yi==1 :
            conditional_sum_x = conditional_sum_x + xi
            y_pos+= 1
    return conditional_sum_x/y_pos


def calculate_mu0(X, Y):
    m = len(Y)
    n=len(X[0])
    y_neg=0
    conditional_sum_x = np.zeros((1,n))
    
    for i in range(0,m) :
        xi = X[i]
        yi = Y[i]
        if yi==0 :
            conditional_sum_x = conditional_sum_x + xi
            y_neg+= 1
    return conditional_sum_x/y_neg

def calculate_prob(x, mu0, mu1, phi, sigma):
        M0=np.matmul(np.matmul(mu0,np.linalg.inv(sigma)),   np.transpose(mu0))
        M1=np.matmul(np.matmul(mu1,np.linalg.inv(sigma)),   np.transpose(mu1))
        theta= 0.5*(np.linalg.det(M0-M1))-np.log(1/phi-1)
        theta_T=np.matmul(mu1-mu0,np.linalg.inv(sigma)) 
        temp=np.matmul(theta_T,np.transpose(x))
        val=temp+theta
        return (1/(1+np.exp(-val)))                                                            
                                                      


def main() :
    print("1st dataset:")
    # Importing 1st dataset
    df_train = pd.read_csv("D:\Python\My files\ds1_train.csv" )
    X = df_train.iloc[:,:-1].values
    Y = df_train.iloc[:,-1:].values
   
    df_test = pd.read_csv("D:\Python\My files\ds1_test.csv")   
    Y_train= np.array(df_train['y'])
    X_train= np.array(df_train.drop('y',axis=1))
    Y_test= np.array(df_test['y'])
    X_test= np.array(df_test.drop('y',axis=1))
    
    val_phi= calculate_phi(Y_train)
    print("Value of phi for training set=", val_phi)
    
    mu0= calculate_mu0(X_train,Y_train)
    mu1= calculate_mu1(X_train,Y_train)
    print("Mu0 and Mu1 on training set=", mu0,mu1)      
    sigma= calculate_sigma(X_train, Y_train, mu0, mu1)
    print("Sigma on training set= ", sigma)      

    prob_train=[]
    for i in range(len(X_train)):
        p=calculate_prob(X_train[i], mu0, mu1, val_phi, sigma)
        if p>0.5 :
            prob_train.append(1)
        else :
            prob_train.append(0)
            
    # measure performance    
    performance = 0    
    # counter    
    count = 0    
    for count in range(len(prob_train))  :  
        
        if Y_train[count] == prob_train[count] :            
            performance = performance + 1
          
        count = count + 1
          
    print( "Accuracy on train set by our first model       :  ", ( performance / count ) * 100 )
    
    val_phi_test= calculate_phi(Y_test)
    print("Value of phi for test set=", val_phi_test)
    mu0_test= calculate_mu0(X_test,Y_test)
    mu1_test= calculate_mu1(X_test,Y_test)
    print("Mu0 and Mu1 on test set=", mu0_test,mu1_test)
    sigma_test= calculate_sigma(X_test, Y_test, mu0_test, mu1_test)
    print("Sigma on test set=", sigma)
    prob_test=[]
    for i in range(len(X_test)):
        p=calculate_prob(X_test[i], mu0_test, mu1_test, val_phi_test, sigma_test)
        if p>0.5 :
            prob_test.append(1)
        else :
            prob_test.append(0)
      
    # measure performance    
    performance = 0    
    # counter    
    count = 0    
    for count in range(len(prob_test))  :  
        
        if Y_test[count] == prob_test[count] :            
            performance = performance + 1
          
        count = count + 1
          
    print( "Accuracy on test set by our first model       :  ", ( performance / count ) * 100 )
    
    print("")
    print("")
    print("")
    print("2nd data set:")
    #importing 2nd dataset
    df_train = pd.read_csv("D:\Python\My files\ds2_train.csv" )
    X = df_train.iloc[:,:-1].values
    Y = df_train.iloc[:,-1:].values
   
    df_test = pd.read_csv("D:\Python\My files\ds2_test.csv")   
    Y_train= np.array(df_train['y'])
    X_train= np.array(df_train.drop('y',axis=1))
    Y_test= np.array(df_test['y'])
    X_test= np.array(df_test.drop('y',axis=1))
    
    val_phi= calculate_phi(Y_train)
    print("Value of phi for training set=", val_phi)
    
    mu0= calculate_mu0(X_train,Y_train)
    mu1= calculate_mu1(X_train,Y_train)
    print("Mu0 and Mu1 on training set=", mu0,mu1)      
    sigma= calculate_sigma(X_train, Y_train, mu0, mu1)
    print("Sigma on training set= ", sigma)      

    prob_train=[]
    for i in range(len(X_train)):
        p=calculate_prob(X_train[i], mu0, mu1, val_phi, sigma)
        if p>0.5 :
            prob_train.append(1)
        else :
            prob_train.append(0)
            
    # measure performance    
    performance = 0    
    # counter    
    count = 0    
    for count in range(len(prob_train))  :  
        
        if Y_train[count] == prob_train[count] :            
            performance = performance + 1
          
        count = count + 1
          
    print( "Accuracy on train set by our second model       :  ", ( performance / count ) * 100 )
    
    val_phi_test= calculate_phi(Y_test)
    print("Value of phi for test set=", val_phi_test)
    mu0_test= calculate_mu0(X_test,Y_test)
    mu1_test= calculate_mu1(X_test,Y_test)
    print("Mu0 and Mu1 on test set=", mu0_test,mu1_test)
    sigma_test= calculate_sigma(X_test, Y_test, mu0_test, mu1_test)
    print("Sigma on test set=", sigma)
    prob_test=[]
    for i in range(len(X_test)):
        p=calculate_prob(X_test[i], mu0_test, mu1_test, val_phi_test, sigma_test)
        if p>0.5 :
            prob_test.append(1)
        else :
            prob_test.append(0)
      
    # measure performance    
    performance = 0    
    # counter    
    count = 0    
    for count in range(len(prob_test))  :  
        
        if Y_test[count] == prob_test[count] :            
            performance = performance + 1
          
        count = count + 1
          
    print( "Accuracy on test set by our second model       :  ", ( performance / count ) * 100 )
    
    
    
    
main()
```
